```python
def read_file(filename):
    data = []
    f = open(filename, "r")
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        if line != "":
            parts = line.split(",")
            data.append(parts)
    f.close()
    return data

def write_file(filename, data_list):
    f = open(filename, "w")
    for record in data_list:
        f.write(",".join(record) + "\n")
    f.close()


# Product Management


def add_product():
    products = read_file("products.txt")
    pid = input("Enter Product ID: ")
    name = input("Enter Product Name: ")
    price = input("Enter Price: ")
    stock = input("Enter Quantity in Stock: ")
    products.append([pid, name, price, stock])
    write_file("products.txt", products)
    print("Product added successfully!\n")

def view_products():
    products = read_file("products.txt")
    print("\n--- Product Catalog ---")
    if len(products) == 0:
        print("No products available.\n")
    else:
        print("ID\tName\t\tPrice\tStock")
        for p in products:
            print(p[0], "\t", p[1], "\t\t", p[2], "\t", p[3])
    print()

def update_product():
    products = read_file("products.txt")
    pid = input("Enter Product ID to update: ")
    found = 0
    for p in products:
        if p[0] == pid:
            found = 1
            p[1] = input("Enter new name: ")
            p[2] = input("Enter new price: ")
            p[3] = input("Enter new stock quantity: ")
    if found == 1:
        write_file("products.txt", products)
        print("Product updated successfully!\n")
    else:
        print("Product not found!\n")

def delete_product():
    products = read_file("products.txt")
    pid = input("Enter Product ID to delete: ")
    new_list = []
    deleted = 0
    for p in products:
        if p[0] != pid:
            new_list.append(p)
        else:
            deleted = 1
    if deleted == 1:
        write_file("products.txt", new_list)
        print("Product deleted successfully!\n")
    else:
        print("Product not found!\n")


# Customer Management

def register_customer():
    customers = read_file("customers.txt")
    cid = input("Enter Customer ID: ")
    name = input("Enter Customer Name: ")
    phone = input("Enter Phone Number: ")
    customers.append([cid, name, phone])
    write_file("customers.txt", customers)
    print("Customer registered successfully!\n")

def view_customers():
    customers = read_file("customers.txt")
    print("\n--- Customer List ---")
    if len(customers) == 0:
        print("No customers found.\n")
    else:
        print("ID\tName\t\tPhone")
        for c in customers:
            print(c[0], "\t", c[1], "\t\t", c[2])
    print()


# Order Placement and Billing


def place_order():
    customers = read_file("customers.txt")
    products = read_file("products.txt")
    orders = read_file("orders.txt")

    cid = input("Enter Customer ID: ")
    customer = None
    for c in customers:
        if c[0] == cid:
            customer = c

    if customer == None:
        print("Customer not found. Register first!\n")
        return

    order_items = []
    total = 0

    while True:
        view_products()
        pid = input("Enter Product ID to buy (or 0 to stop): ")
        if pid == "0":
            break

        quantity = int(input("Enter quantity: "))
        found = 0

        for p in products:
            if p[0] == pid:
                found = 1
                stock = int(p[3])
                if quantity <= stock:
                    amount = int(p[2]) * quantity
                    total = total + amount
                    order_items.append([pid, p[1], str(quantity), str(amount)])
                    p[3] = str(stock - quantity)
                else:
                    print("Not enough stock available!")
        if found == 0:
            print("Invalid Product ID!")

    if len(order_items) == 0:
        print("No items selected. Order cancelled.\n")
        return

    write_file("products.txt", products)

    order_id = "ORD" + str(len(orders) + 1)
    orders.append([order_id, cid, str(total)])
    write_file("orders.txt", orders)

    print("\n--- Bill ---")
    print("Customer:", customer[1], "(", customer[2], ")")
    print("Order ID:", order_id)
    print("Items Purchased:")
    for item in order_items:
        print(item[1], "x", item[2], "=", "₹" + item[3])
    print("Total Amount: ₹" + str(total))
    print("Order placed successfully!\n")


# Menus

def admin_menu():
    while True:
        print("=== ADMIN MENU ===")
        print("1. Add Product")
        print("2. View Products")
        print("3. Update Product")
        print("4. Delete Product")
        print("5. View Customers")
        print("6. Back to Main Menu")
        ch = input("Enter choice: ")

        if ch == "1":
            add_product()
        elif ch == "2":
            view_products()
        elif ch == "3":
            update_product()
        elif ch == "4":
            delete_product()
        elif ch == "5":
            view_customers()
        elif ch == "6":
            break
        else:
            print("Invalid choice!\n")

def customer_menu():
    while True:
        print("=== CUSTOMER MENU ===")
        print("1. Register")
        print("2. View Products")
        print("3. Place Order")
        print("4. Back to Main Menu")
        ch = input("Enter choice: ")

        if ch == "1":
            register_customer()
        elif ch == "2":
            view_products()
        elif ch == "3":
            place_order()
        elif ch == "4":
            break
        else:
            print("Invalid choice!\n")

def main_menu():
    while True:
        print("=== MINI E-COMMERCE SYSTEM ===")
        print("1. Admin")
        print("2. Customer")
        print("3. Exit")
        ch = input("Enter choice: ")

        if ch == "1":
            admin_menu()
        elif ch == "2":
            customer_menu()
        elif ch == "3":
            print("Thank you for using the system!")
            break
        else:
            print("Invalid choice!\n")

main_menu()
```
